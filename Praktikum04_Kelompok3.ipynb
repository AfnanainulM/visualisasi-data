##### **B. Kustomisasi Basic Bar Chart**

```python
import matplotlib.pyplot as plt

# data
jurusan = [
    'Ilmu Komputer',
    'Teknik Informatika',
    'Sistem Informasi',
    'Data Sience'
]

jumlah_mahasiswa = [120, 150, 100, 80]

# membuat baris bar chart
plt.bar(jurusan, jumlah_mahasiswa, color=['blue', 'orange', 'green', 'red'])
plt.xlabel('Jurusan')
plt.ylabel('Jumlah Mahasiswa')
plt.xticks(rotation=45)

# menambahkan label pada batang
for i, v in enumerate(jumlah_mahasiswa):
    plt.text(i, v + 2, str(v), ha='center')

plt.title('Jumlah Mahasiswa Per Jurusan')
plt.show()
```

##### **C. Multiple Bar Chart**




```python
import matplotlib.pyplot as plt

# data
jurusan = [
    'Ilmu Komputer',
    'Teknik Informatika',
    'Sistem Informasi',
    'Data Sience'
]

jumlah_mahasiswa = [120, 150, 100, 80]
jumlah_mahasiswa_2024 = [110, 140, 95, 85]

x = range(len(jurusan))
width = 0.4

plt.bar(x, jumlah_mahasiswa, width, label='2023', color='orange')
plt.bar([p + width for p in x], jumlah_mahasiswa_2024, width,
        label='2024', color='skyblue')

# membuat baris bar chart
plt.title('Jumlah Mahasiswa Per Jurusan (2023 vs 2024)')
plt.xlabel('Jurusan')
plt.ylabel('Jumlah Mahasiswa')
plt.xticks(rotation=45)
plt.xticks([p + width/2 for p in x], jurusan)
plt.legend()

for i, v in enumerate(jumlah_mahasiswa):
    plt.text(i, v + 2, str(v), ha='center')

plt.show()
```

##### **D. Basic Bar Chart untuk Matplotlib**

```python
import matplotlib.pyplot as plt

# data
jurusan = ['Ilmu Komputer', 'Teknik Informatika', 'Sistem Informasi', 'Data Sience']
tahun = ['2019', '2020', '2021', '2022', '2023']
data = {
    'Ilmu Komputer' : [100, 110, 120, 130, 140],
    'Teknik Informatika' : [120, 120, 132, 145, 160],
    'Sistem Informasi' : [90, 95, 100, 105, 110],
    'Data Sience' : [70, 75, 80, 85, 90]
}
width = 0.2
x = range(len(tahun))

# visualisasi masing-masing jurusan
plt.bar(x, data['Ilmu Komputer'], width=width, label='Ilmu Komputer', color='blue')
plt.bar([p + width for p in x],
        data['Teknik Informatika'], width=width, label='Teknik Informatika', color='green')
plt.bar([p + width * 2 for p in x],
        data['Sistem Informasi'], width=width, label='Sistem Informasi', color='orange')
plt.bar([p + width * 3 for p in x],
        data['Data Sience'], width=width, label='Data Sience', color='red')

# detail grafik
plt.title('Tren Jumlah Mahasiswa Per Jurusan (2019-2023)')
plt.xlabel('Tahun')
plt.ylabel('Jumlah Mahasiswa')
plt.xticks([p + width for p in x], tahun)
plt.legend()
plt.show()
```
